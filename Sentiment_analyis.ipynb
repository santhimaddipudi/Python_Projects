```python
import pandas as pd
df=pd.read_csv('/content/sample_data/influencers_data.csv')
df.head(1)
```

```python
from textblob import TextBlob
#calculate sentiment polarity
df['polarity']=df['content'].apply(lambda x:TextBlob(str(x)).sentiment.polarity)
df['Sentiment']=df['polarity'].apply(lambda x:'positive'if x>0
                                     else('neutral' if x==0 else 'negative'))
df[['content','Sentiment']].head(4)
```
